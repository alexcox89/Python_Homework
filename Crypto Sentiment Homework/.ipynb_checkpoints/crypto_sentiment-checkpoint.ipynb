# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Reading the News API key enviroment variable
abs_path = Path(r'C:/Users/Alex/Desktop/.env')
load_dotenv(abs_path) 
api_key=os.getenv('NEWS_API_KEY')
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
# For this, I referenced Unit 12, #2, activity 07-Stu_Crisis_Voice
btc_headlines = newsapi.get_everything(q="bitcoin", language="en")
btc_headlines['totalResults']
```

```python
# Fetch the Ethereum news articles
eth_headlines = newsapi.get_everything(
    q="ethereum",
    language="en"
)
eth_headlines["totalResults"]
```

```python
# Create the Bitcoin sentiment scores DataFrame
# This code is taken from the Vader Sentiment Exercise in Unit 12, 2 
btc_sentiments = []

for article in btc_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        btc_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
btc = pd.DataFrame(btc_sentiments)
```

```python
btc = btc.drop(columns=["date"])
```

```python
btc
```

```python
# Create the ethereum sentiment scores DataFrame
eth_sentiments = []

for articles in eth_headlines["articles"]:
    try:
        text = articles["content"]
        date = article["publishedAt"][:10]
        results = analyzer.polarity_scores(text)
        compound = results["compound"]
        pos = results["pos"]
        neu = results["neu"]
        neg = results["neg"]

        eth_sentiments.append({
            "text": text,
            "date": date,
            "Compound": compound,
            "Positive": pos,
            "Negative": neg,
            "Neutral": neu,
        })
    except AttributeError:
        pass
    
eth  = pd.DataFrame(eth_sentiments)
```

```python
eth = eth.drop(columns=["date"])
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
# For this section I used a combination of exercises from Unit12-1 to derive the solution
# First we set a variable for the lemmatizer
wnl = WordNetLemmatizer() 
# Next we set a variable for the stopwords
sw = set(stopwords.words('english'))
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    #clean up text
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    
    # Create a list of the words from the cleaned text
    words = word_tokenize(re_clean)
    
    # Lemmatize Words into root words
    lem = [wnl.lemmatize(word) for word in words]

    # Convert the words to lowercase and remove stopwords
    tokens = [word.lower() for word in lem if word.lower() not in sw]

    return tokens
```

```python
# Create a new tokens column for bitcoin
btc["tokens"] = btc.text.apply(tokenizer)
btc
```

```python
# Create a new tokens column for ethereum
eth["tokens"] = eth.text.apply(tokenizer)
eth.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
N = 2
# We concatenate all of the strings in the series in order to provide the total text
text1 = btc.text.str.cat()
# We want the ngrams of the tokenized texts, so we apply the tokenizer from above
Ngrams = ngrams(tokenizer(text1), N)
Counter(Ngrams).most_common(20)
# Alternate solution tried:
#def bigram_counter(tokens): 
    #for words in btc['tokens']:
       # bigrams = ngrams(words, n=2)
       # top_10 = dict(Counter(bigrams).most_common(10))
   # return pd.DataFrame(list(top_10.items()), columns=['bigram', 'count'])
```

```python
# Generate the Ethereum N-grams where N=2
N = 2
# We concat each list of texts in the series
text2 = eth.text.str.cat()
# We get the ngrams of the tokenized text
Ngrams = ngrams(tokenizer(text2), N)
Counter(Ngrams).most_common(20)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
btc_tokens = tokenizer(text1)
token_count(btc_tokens, 10)
```

```python
# Get the top 10 words for Ethereum
eth_tokens = tokenizer(text2)
token_count(eth_tokens, 10)
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
# The basis of this code is exercise 04-Stu_NER_Clouds in Unit 12-3
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
# We did this above and defined it as text1
text1
```

```python
# Run the NER processor on all of the text
doc = nlp(text1)

# Add a title to the document
doc.user_data["title"] = "Bitcoin NER"
```

```python
# Render the visualization
displacy.render(doc, style='ent')
```

```python
# List all Entities
for ent in doc.ents:
    print(ent.text, ent.label_)
```
