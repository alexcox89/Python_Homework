```python
##Import libraries and setup Plaid key variables
```

```python
import os
import plaid
import requests
import datetime
import json
import pandas as pd
from dotenv import load_dotenv

%matplotlib inline
```

```python
load_dotenv()
```

```python
PLAID_CLIENT_ID = os.getenv("PLAID_CLIENT_ID")
PLAID_SECRET = os.getenv("PLAID_SECRET")
PLAID_PUBLIC_KEY = os.getenv("PLAID_PUBLIC_KEY")
```

```python
###Create clients and get tokens
```

```python
client = plaid.Client(client_id=PLAID_CLIENT_ID, secret=PLAID_SECRET, public_key=PLAID_PUBLIC_KEY, environment="sandbox",) 
```

```python
INSITUTION_ID = "ins_109508"
```

```python
create_response = client.Sandbox.public_token.create(INSITUTION_ID, ["transactions", "income", "assets"])
```

```python
create_response
```

```python
get_access_token = client.Item.public_token.exchange(create_response["public_token"])
```

```python
get_access_token
```

```python
access_token = get_access_token["access_token"]
```

```python
###Get Accounts using access_token
```

```python
accounts = client.Accounts.get(access_token)
```

```python
accounts
```

```python
##In the next section, we will acquire the transactions from the last 90 days and put the transaction data into a dataframe
```

```python
### Set start and end dates and retrieve transactions
```

```python
start_date = "{:%Y-%m-%d}".format(datetime.datetime.now() + datetime.timedelta(-90))
end_date = "{:%Y-%m-%d}".format(datetime.datetime.now())
```

```python
transactions_response = client.Transactions.get(access_token, start_date, end_date)
```

```python
transactions_response
```

```python
###Get the categories for each transaction
```

```python
for transaction in transactions_response["transactions"]:
    print(transaction["category"])
```

```python
### Create Dataframe
```

```python
transaction_df = pd.DataFrame(columns=["date", "name", "amount", "category"])
```

```python
transactions = transactions_response["transactions"]
```

```python
for index, transaction in enumerate(transactions):
    transaction_df.loc[index] = [transaction["date"], transaction["name"], transaction["amount"], transaction["category"][0]]
```

```python
transaction_df.head()
```

```python
transaction_df.dtypes
```

```python
### Set data to their appropriate types
```

```python
transaction_df["amount"] = transaction_df["amount"].astype("float")
```

```python
transaction_df["date"] = pd.to_datetime(transaction_df["date"], format="%Y-%m-%d")
```

```python
transaction_df.dtypes
```

```python
### Set the index as the date column
transaction_df = transaction_df.set_index("date")
```

```python
transaction_df.head()
```

```python
## In this section, we get income data using the client and access token and access the arrays of data
```

```python
income_data = client.Income.get(access_token)
```

```python
income_data
```

```python
print(f"Last Year's income: {income_data['income']['last_year_income_before_tax']}")
```

```python
print(f"Current monthly income: {income_data['income']['income_streams'][0]['monthly_income']}")
```

```python
print(f"Projected Year's income: {income_data['income']['projected_yearly_income_before_tax']}")
```

```python
## In this final section we will perform budget analysis by category and per month.
```

```python
totals_by_category = transaction_df.groupby("category").sum()["amount"]
```

```python
totals_by_category
```

```python
totals_by_category.plot.pie(title="Total Expenses Per Category")
```

```python
transactions_per_month = transaction_df.reset_index()
```

```python
transactions_per_month["month"] = pd.DatetimeIndex(transactions_per_month["date"]).month
```

```python
transactions_per_month = transactions_per_month.groupby("month").sum()
```
